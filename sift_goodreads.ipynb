```python
import pandas
goodreads = pandas.read_csv('goodreads_library_export.csv')
goodreads.columns
```

We're only interested in a few columns.

```python
thing = goodreads[['Title', 'Average Rating', 'Author', 'Bookshelves', 'Bookshelves with positions', 'Number of Pages', 'Year Published']]
```

```python
thing.Bookshelves.unique()
```

```python
unread = thing[thing.Bookshelves.isin(['to-read', 'currently-reading'])]
```

```python
items = pandas.DataFrame(unread)
```

```python
items['order'] = pandas.to_numeric(items['Bookshelves with positions'].str.extract('#([0-9]*)'))
```

```python
items.sort_values(by='order')
```
